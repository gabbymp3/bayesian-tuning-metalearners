```python
# Grid search
def tune_grid_search(X, W, Y, tau, base_estimator, param_grid, cv=5, n_jobs=-1, verbose=1):
    yw = np.column_stack((Y, W))
    estimator = EconMLWrapper(est=base_estimator)
    grid = GridSearchCV(
        estimator=estimator,
        param_grid=param_grid,
        cv=cv,
        n_jobs=n_jobs,
        verbose=verbose,
    )
    grid.fit(X, yw, score_y=tau)
    best = grid.best_estimator_
    tau_hat = best.predict(X)
    pehe = -np.mean((tau - tau_hat) ** 2)

    return grid.best_estimator_.est
    
```
